# Exploratory Data Analysis with YData Profiling
This notebook generates an Exploratory Data Analysis (EDA) report for each CSV file in the `pre-combine` directory using the `ydata-profiling` library (formerly `pandas-profiling`).

**Note:** Ensure you have the library installed by running `pip install ydata-profiling`.

```python
import pandas as pd
import os
import glob
from ydata_profiling import ProfileReport

# Set the directory containing the CSV files
data_dir = './'
files = glob.glob(os.path.join(data_dir, '*.csv'))
print(f'Found {len(files)} CSV files.')
```

```python
output_dir = 'yprofile_reports'
os.makedirs(output_dir, exist_ok=True)

for f in files:
    file_name = os.path.basename(f)
    print(f'\nGenerating profile for {file_name}...')
    try:
        df = pd.read_csv(f)
        # Generate report
        profile = ProfileReport(df, title=f'{file_name} Profiling Report', explorative=True)
        
        # Save report as HTML
        output_path = os.path.join(output_dir, file_name.replace('.csv', '_report.html'))
        profile.to_file(output_path)
        print(f'Saved report to {output_path}')
    except Exception as e:
        print(f'Error generating profile for {file_name}: {e}')

print('\nAll profiles generated successfully.')
```
